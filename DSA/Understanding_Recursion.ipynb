# Understanding Recursion

### What is Recursion?
Recursion is a process where a function directly or indirectly calls itself, and the function involved in this process is termed a recursive function. It's a technique used to solve certain problems effectively, such as Towers of Hanoi, Tree Traversals (Inorder/Preorder/Postorder), and Graph Depth-First Search (DFS). A recursive function addresses a problem by invoking copies of itself, solving smaller subproblems of the original issue. Multiple recursive calls can be generated as needed. However, it's crucial to provide a stopping condition to prevent infinite recursion. In essence, each function call simplifies the original problem.

### **The Need for Recursion**
Recursion is a powerful approach that streamlines code, enhancing its readability and writability. It offers advantages over iterative techniques, which we'll discuss later. Whenever a task can be broken down into similar subtasks, recursion emerges as an excellent solution strategy. For instance, calculating the factorial of a number.

### **Key Characteristics of Recursion:**

Repeating the same operations with varying inputs.
Progressively using smaller inputs to reduce the problem.
A base condition is essential to halt the recursion to prevent infinite looping.





### **Memory Storage of Recursive Functions**
Recursion utilizes more memory because each recursive call adds to the stack and retains values until the call completes. This process adheres to the Last In First Out (LIFO) structure akin to a stack data structure.

### Understanding Base Conditions in Recursion
In recursive programming, the base case provides a solution for the simplest scenario, and the solution for larger problems is expressed in terms of smaller ones. For instance, in the code snippet below:

```python
def fact(n):
    if n <= 1:  # Base case
        return 1
    else:
        return n * fact(n-1)
```

Here, the base case for n1 is defined, and any larger value is tackled by breaking it down into smaller values until reaching the base case.

**Solving Problems through Recursion**
The concept involves representing a problem in relation to one or more smaller instances of itself and introducing one or more base conditions that halt the recursion. For instance, to compute the factorial of n, one relies on the factorial of n-1=0, with n=0 serving as the base case, returning 1.

**Memory Allocation for Recursive Function Calls**
When a function is called, memory is allocated on the stack. In recursion, a function calls itself, and the memory for each called function is allocated on top of the memory allocated to the calling function. This results in separate instances of local variables for each function call. Upon reaching the base case, the function returns its value to the caller, memory is released, and the process continues.

Recursion Illustrated
Consider the following example:

```python
def printFun(test):
    if test < 1:
        return
    else:
        print(test, end=" ")
        printFun(test-1)  # Recursive call
        print(test, end=" ")
        return

test = 3
printFun(test)
```

The output will be: 3 2 1 1 2 3
In this case, when printFun(3) is called, memory is allocated to it, and subsequent recursive calls create new instances with separate variables. This results in the pattern seen in the output.

### **Recursion vs. Iteration**


**Recursion:**

- Terminates with the base case becoming true.
- Applicable to functions.
- Consumes more memory due to stack usage with each recursive call.
- Leads to smaller code size.


**Iteration:**

- Terminates when the condition becomes false.
- Employed with loops.
- Requires minimal extra memory.
- Can result in larger code size.


**Summarizing Recursion:**

- Involves recursive and base cases.
- Base case terminates the recursion.
- Each recursive call has its own memory space.
- Infinite recursion can exhaust stack memory.

### Examples of recursive algorithms include Merge Sort, Quick Sort, Tower of Hanoi, Fibonacci Series, and Factorial Problem.

## 1. Print 1 to N using Recursion

```python
def print1toN(n):
    """
    Prints numbers from 1 to N using recursion.

    This function uses recursion to print numbers from 1 to N in ascending order.
    It starts with the base case where N equals 0, and then recursively calls
    itself for values from N-1 down to 1.

    Args:
        n (int): The upper limit of the range to print.

    Returns:
        None
    """
    if n == 0:
        return  # Base case: Stop recursion when n reaches 0
    print1toN(n - 1)  # Recursive call with n-1
    print(n, end=" ")  # Print the current value of n


# Example usage
print1toN(3)  # Output: 1 2 3
```

## 2. Print N to 1 using Recursion in Python

```python
def PrintReverseOrder(N):
    """
    Prints numbers from N to 1 in reverse order using recursion.

    This function uses recursion to print numbers from N down to 1 in reverse order.
    It starts with the base case where N is less than or equal to 0, and then recursively
    calls itself with the value N - 1.

    Args:
        N (int): The starting value for the reverse order range.

    Returns:
        None
    """
    if N <= 0:
        return  # Base case: Stop recursion when N is less than or equal to 0
    else:
        print(N, end=" ")  # Print the current value of N
        PrintReverseOrder(N - 1)  # Recursive call with N - 1

# Example usage
N = 5
PrintReverseOrder(N)  # Output: 5 4 3 2 1
```

## 3. Sum of Digits using Recursion



```python
def sum_of_digit(n):
    """
    Calculates the sum of digits of a number using recursion.

    This function uses recursion to calculate the sum of digits of a given number 'n'.
    It starts by checking if 'n' is equal to 0, which serves as the base case to stop
    the recursion. If 'n' is not 0, it calculates the last digit using n % 10, and then
    makes a recursive call with the remaining digits by dividing n by 10.

    Args:
        n (int): The number for which the sum of digits needs to be calculated.

    Returns:
        int: The sum of digits of the given number.
    """
    if n == 0:
        return 0  # Base case: Stop recursion when n is 0
    return (n % 10 + sum_of_digit(int(n / 10)))  # Calculate sum of current digit and recursive call

# Example usage
num = 12345
result = sum_of_digit(num)
print("Sum of digits in", num, "is", result)  # Output: Sum of digits in 12345 is 15
```

## 4. Palindrome check using Recursion

Time Complexity: O(n)


Auxiliary Space: O(n)

```python

def isPalRec(st, s, e):
    """
    Recursive function to check if a substring in a given string is palindrome.

    This function checks whether the substring st[s..e] is a palindrome or not.
    It starts by comparing the first and last characters of the substring. If they
    match, it recursively calls itself for the remaining substring between s+1 and e-1.

    Args:
        st (str): The input string.
        s (int): The starting index of the current substring.
        e (int): The ending index of the current substring.

    Returns:
        bool: True if the substring is a palindrome, False otherwise.
    """
    # If there is only one character
    if s == e:
        return True

    # If first and last characters do not match
    if st[s] != st[e]:
        return False

    # If there are more than two characters,
    # check if middle substring is also palindrome
    if s < e + 1:
        return isPalRec(st, s + 1, e - 1)

    return True

def isPalindrome(st):
    """
    Checks if a given string is a palindrome using recursion.

    This function uses the isPalRec function to check whether the entire string
    is a palindrome or not.

    Args:
        st (str): The input string.

    Returns:
        bool: True if the string is a palindrome, False otherwise.
    """
    n = len(st)

    # An empty string is considered as palindrome
    if n == 0:
        return True

    return isPalRec(st, 0, n - 1)

# Example usage
st = "geeg"
if isPalindrome(st):
    print("Yes")  # Output: Yes
else:
    print("No")

```

## 5. Sum of natural numbers using recursion

Time complexity : O(n)

Auxiliary space : O(n)

```python
def recurSum(n):
    """
    Calculates the sum of the first 'n' natural numbers using recursion.

    This function recursively calculates the sum of the first 'n' natural numbers.
    It starts by checking if 'n' is less than or equal to 1, which serves as the base
    case to stop the recursion. If 'n' is not 1, it adds 'n' to the sum of the first
    'n-1' natural numbers, calculated by making a recursive call.

    Args:
        n (int): The number of natural numbers to sum.

    Returns:
        int: The sum of the first 'n' natural numbers.
    """
    if n <= 1:
        return n  # Base case: Stop recursion when n is 1 or less
    return n + recurSum(n - 1)  # Add n to the sum of the first n-1 natural numbers

# Example usage
n = 5
print(recurSum(n))  # Output: 15 (1 + 2 + 3 + 4 + 5)
```

## 6. Count Total Digits in a Number


```python
def countDigits(n):
    """
    Counts the number of digits in a given number using recursion.

    This function calculates the count of digits in a given number 'n' using recursion.
    It starts by checking if 'n' is less than 10, which serves as the base case to stop
    the recursion. If 'n' is not less than 10, it adds 1 to the count of digits and makes
    a recursive call with the number obtained by integer division of 'n' by 10.

    Args:
        n (int): The given number.

    Returns:
        int: The count of digits in the given number 'n'.
    """
    if n < 10:
        return 1  # Base case: Stop recursion when n is less than 10
    return (1 + countDigits(n // 10))  # Add 1 to count and make recursive call

# Example usage
n = 12345
count = countDigits(n)
print(f"Number of digits in {n} is {count}")  # Output: Number of digits in 12345 is 5
```

## 7. Fibonacci using Recursion

```python
def fibonacci(n):
    """
    Calculates the nth Fibonacci number using recursion.

    This function calculates the nth Fibonacci number using recursion.
    It starts by checking if 'n' is equal to 1 or 2, which serves as
    the base cases to stop the recursion. If 'n' is neither 1 nor 2, it
    calculates the nth Fibonacci number by summing up the (n-1)th and (n-2)th
    Fibonacci numbers.

    Args:
        n (int): The index of the desired Fibonacci number.

    Returns:
        int: The nth Fibonacci number.
    """
    if n == 1:
        return 1  # Base case: Fibonacci(1) is 1
    if n == 2:
        return 1  # Base case: Fibonacci(2) is 1
    return fibonacci(n - 1) + fibonacci(n - 2)  # Sum of (n-1)th and (n-2)th Fibonacci numbers

# Example usage
n = 7
result = fibonacci(n)
print(f"The {n}th Fibonacci number is {result}")  # Output: The 7th Fibonacci number is 13
```

## 8. Factorial using Recursion

```python
def factorial(n):
    """
    Calculates the factorial of a number 'n' using recursion.

    This function calculates the factorial of a given number 'n' using recursion.
    It starts by checking if 'n' is equal to 0 or 1, both of which serve as base cases
    to stop the recursion. If 'n' is neither 0 nor 1, it calculates the factorial by
    multiplying 'n' with the factorial of (n-1), calculated by making a recursive call.

    Args:
        n (int): The given number for which the factorial is to be calculated.

    Returns:
        int: The factorial of the given number 'n'.
    """
    if n == 0:
        return 1  # Base case: Factorial of 0 is 1
    if n == 1:
        return 1  # Base case: Factorial of 1 is 1
    return n * factorial(n - 1)  # Calculate factorial of n by multiplying with factorial of (n-1)

# Example usage
n = 5
result = factorial(n)
print(f"The factorial of {n} is {result}")  # Output: The factorial of 5 is 120
```

## 9. Find nCr


```python
def nCr(n, r):
    """
    Calculates the value of nCr (binomial coefficient) using recursion.

    This function calculates the value of nCr (n choose r) using recursion.
    It starts by checking if 'n' is 0, which serves as the base case to stop the
    recursion. If 'r' is equal to 0 or 'r' is equal to 'n', both of these are base
    cases that return 1. Otherwise, the function calculates nCr by summing up the
    values of nCr(n-1, r) and nCr(n-1, r-1), making recursive calls to compute these
    values.

    Args:
        n (int): The total number of items.
        r (int): The number of items to choose.

    Returns:
        int: The value of nCr (binomial coefficient).
    """
    if n == 0:
        return 0  # Base case: Stop recursion when n is 0

    if r == n or r == 0:
        return 1  # Base cases: nCr when r is 0 or r is equal to n is 1

    return nCr(n - 1, r) + nCr(n - 1, r - 1)  # Calculate nCr by recursive calls

# Example usage
n = 5
r = 2
result = nCr(n, r)
print(f"{n}C{r} is {result}")  # Output: 5C2 is 10
```

## 10. Find GCD

```python
def GCD(a, b):
    """
    Calculates the Greatest Common Divisor (GCD) of two numbers using recursion.

    This function calculates the GCD of two given numbers 'a' and 'b' using recursion.
    It starts by checking if 'b' is 0, which serves as the base case to stop the recursion.
    If 'b' is not 0, the function calculates the GCD using the property that GCD(a, b) is
    equal to GCD(b, a % b). It makes a recursive call with 'b' and 'a % b' to calculate
    the GCD.

    Args:
        a (int): The first number.
        b (int): The second number.

    Returns:
        int: The Greatest Common Divisor (GCD) of the two numbers.
    """
    if b == 0:
        return a  # Base case: Stop recursion when b is 0
    return GCD(b, a % b)  # Calculate GCD using GCD property and recursive call

# Example usage
a = 48
b = 18
result = GCD(a, b)
print(f"The GCD of {a} and {b} is {result}")  # Output: The GCD of 48 and 18 is 6
```
